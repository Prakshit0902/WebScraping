```python
from bs4 import BeautifulSoup
import pandas as pd
import numpy as np
import requests
```

```python
headers={'User-Agent':'Mozilla/5.0 (Windows NT 6.3; Win 64 ; x64) Apple WeKit /537.36(KHTML , like Gecko) Chrome/132.0.6834.160 Safari/537.36'}
webpage = requests.get('https://www.espncricinfo.com/records/season/team-match-results/2011-2011?trophy=117',headers=headers).text
```

```python
soup = BeautifulSoup(webpage,'lxml')
table = soup.find('table')
table
```

```python
columns = []
tr = table.find_all('div',class_ = 'ds-popper-wrapper')
for i in tr:
    columns.append(i.text)
columns
```

```python
rows = []
tbody = table.find('tbody')
for row in tbody.find_all('tr') :
    cols = row.find_all('td')
    cols = [col.text.strip() for col in cols]
    rows.append(cols)
rows

```

```python
df = pd.DataFrame(rows,columns=columns)
df
```

```python
years = ['2007to08-2007to08','2009-2009','2009to10-2009to10','2011-2011','2012-2012','2013-2013','2014-2014','2015-2015','2016-2016','2017-2017','2018-2018','2019-2019','2020to21-2020to21','2021-2021','2022-2022','2023-2023','2024-2024']
columns = ['Team 1', 'Team 2', 'Winner', 'Margin', 'Ground', 'Match Date', 'Scorecard']
rows = []
for i in range(len(years)):
    url = f'https://www.espncricinfo.com/records/season/team-match-results/{years[i]}?trophy=117'
    headers={'User-Agent':'Mozilla/5.0 (Windows NT 6.3; Win 64 ; x64) Apple WeKit /537.36(KHTML , like Gecko) Chrome/132.0.6834.160 Safari/537.36'}
    web = requests.get(url = url,headers=headers).text
    
    soup = BeautifulSoup(web,'lxml')
    
    table = soup.find('table')
    tbody = table.find('tbody')
    
    for row in tbody.find_all('tr') :
        cols = row.find_all('td')
        cols = [col.text.strip() for col in cols]
        rows.append(cols)

df = pd.DataFrame(rows,columns=columns)
df
```

```python
df.to_csv('web_scrapped_cricket_data')
```
